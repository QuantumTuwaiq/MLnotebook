<h1>Problem:</h1>
<p style = "font-size: 25px">We want to Classify written digits from 0 to 9</p>

<h1>Data Preparation:</h1>
<p style = "font-size: 25px">We found a database called digits that has written digits, and the classification of them</p>

```python
import matplotlib.pyplot as plt
# Import datasets, classifiers and performance metrics
from sklearn import datasets, metrics, svm #Data to train the model on, Metrics to check accuracy, SVM ==> Support Vector Machine
from sklearn.model_selection import train_test_split
```

 <p style="font-size: 25px">Let's print a few samples to see how it looks!</p>

```python
digits = datasets.load_digits()

_, axes = plt.subplots(nrows=1, ncols=21, figsize=(10, 3))
for ax, image, label in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(label)
```


  <img src ="Data1.png">

```python
data = digits.images.reshape((len(digits.images), -1))
```

<h1>Creating the model</h1>
<p style="font-size: 25px">We can see that before training, the classifier doesn't fit the data</p>

```python
# Create a classifier: a support vector classifier
classifier = svm.SVC(gamma=0.001)

# Split data into 50% train and 50% test subsets
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.5, shuffle=False
)


```

<img src="Data2.png">

<h1>After training:</h1>
<p style = "font-size: 25px">After the fit() function, the classifier has been fitted to the data and will be more accurate</p>

```python
# Learn the digits on the train subset
classifier.fit(X_train, y_train)

# Predict the value of the digit on the test subset "Ineference"
predicted = classifier.predict(X_test)
```

<img src = "Data3.png">

```python
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for i, image, prediction in zip(range(100), X_test, predicted):
    image = image.reshape(8,8)
    plt.subplot(10,10,i+1)
    plt.ylabel(prediction)
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")

```
